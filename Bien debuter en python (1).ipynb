Fonction de Conversion de Température
Objectif : Écrire une fonction qui convertit la température de Celsius en Fahrenheit.

 Instructions :

Créez une fonction nommée celsius_en_fahrenheit qui prend un nombre (température en degrés Celsius) comme argument.
La fonction doit calculer et retourner la température en degrés Fahrenheit en utilisant la formule :


3:Testez cette fonction en convertissant plusieurs températures de Celsius en Fahrenheit et affichez les résultats.

```python
def celsius_to_fahrenheit(celsius):
    fahrenheit=(celsius*9/5)+32
    return fahrenheit
celsius_to_fahrenheit(2)
```
